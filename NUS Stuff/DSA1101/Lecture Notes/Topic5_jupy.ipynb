```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

play_decision = pd.read_csv("DTdata.csv") 
print(play_decision)
play_decision = pd.get_dummies(play_decision)
print(play_decision.head())

# featmap = {k:v for v, k in enumerate(list(play_decision.columns))}
# print(featmap)
X = play_decision.drop(columns=["Play_no", "Play_yes"], axis=1)
# print(X)
y = play_decision.iloc[:, play_decision.columns == "Play_yes"]
# print(y)
dtree = DecisionTreeClassifier()
dtree = dtree.fit(X, y)

tree.plot_tree(dtree)


```

```python
newdata = X.copy().iloc[0]
newdata.at["Temperature_mild"] = 1
newdata.at["Humidity_high"] = 1
newdata.at["Humidity_normal"] = 0
# print(newdata)

dtree.predict([newdata])
```
